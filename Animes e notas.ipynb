```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
animedf = pd.read_csv('anime.csv', sep=',')
```

```python
# Copiando para um novo dataset para manter as caracteristicas do original
dados = animedf.copy()
dados
```

```python
dados.head()
```

```python
dados.info()
```

```python
nulos_rating = dados['rating'].isnull()
dados[nulos_rating]
```

```python
nulos_genre = dados['genre'].isnull()
dados[nulos_genre]
```

```python
nulos_type = dados['type'].isnull()
dados[nulos_type]
```

### Nas celulas acima podemos ver que existem alguns valores nulos em algumas colunas, em especial a coluna type onde diz a respeitos do formato dos animes, se é de epsisódio ou filme etc. Essa coluna tem poucos valores nulos e conseguimos fazer uma busca na internet a repeitos do formato e identificamos que apenas 1 anime é filme e o restante é tv, então vamos fazer as devidas alterações.


```python
dados.iat[10900, 3] = 'Movie' # Alterando pelo indice do anime 
```

```python
dados[dados.index==10900] # Verificando se a alteração foi feita com sucesso 
```

```python
dados.loc[dados.type.isnull(), 'type'] = 'TV'
```

```python
dados.info()
```

### Agora todos os dados da coluna type estão preenchidos corretamente, as outras colunas como genre que é o gênero do anime e rating que é a nota, nós não temos informações do que poderíamos colocar e simplesmente preencher com a média ou moda pode impactar negativamente na nossa análise dos dados, por hora vamos deixar do jeito que está. 

### Se notarmos bem a coluna episodes está como object, o interessante é essa coluna estivesse como int  já que é numérica, mas como alguns valores desconhecidos estão com o valor 'Unknown' decidi alterara-los para 0 e assim podemos fazer a conversão para int.

```python
dados.loc[dados.episodes == 'Unknown', 'episodes'] = 0
```

```python
dados[nulos_type]
```

```python
dados['episodes'] = dados['episodes'].astype(int)
dados.info()
```

### Os animes normalmente tem formato, uma questão que fica agora é qual a quantidade de animes feitos em relação aos formatos? a coluna 'type' irá nos dizer isso.

```python
formato = pd.DataFrame(dados.type.value_counts())
formato
```

```python
plt.figure(figsize=(15,5))
plt.bar(formato.index, formato.type)
plt.title('Formato dos animes')
plt.xticks(rotation=90)
plt.show()
```

### Após visualizar os formatos, quais será as 10 maiores notas de animes e seus formatos ?

```python
maiores_notas = dados.nlargest(10, 'rating')
maiores_notas
```

```python

plt.figure(figsize=(15,5))
plt.bar(maiores_notas.name, maiores_notas.rating)
plt.title('Animes com as maiores notas')
plt.xticks(rotation=90)
plt.show()
```

### Dos 10 animes com as maiores notas 4 são no formato de filme "Movie", 5 em formato para "TV" e apenas um em OVA.

### E os gêneros com as maiores notas?

```python
# Genero dos animes com as maiores notas
plt.figure(figsize=(15,5))
plt.bar(maiores_notas.genre, maiores_notas.rating)
plt.title('Gênero dos animes com as maiores notas')
plt.xticks(rotation=90)
plt.show()
```

### E Quais são os animes com as notas mais baixas?

```python
menores_notas = dados.nsmallest(10, 'rating')
menores_notas
```

```python

plt.figure(figsize=(15,5))
plt.bar(menores_notas.name, menores_notas.rating)
plt.title('Animes com as menores notas')
plt.xticks(rotation=90)
plt.show()
```

### E os gêneros com as menores notas

```python
plt.figure(figsize=(15,5))
plt.bar(menores_notas.genre, menores_notas.rating)
plt.title('Gênero dos animes com as menores notas')
plt.xticks(rotation=90)
plt.show()
```

### Tendo em vista que os animes possui gêneros e subgêneros, o que dificulta determinar se um gênero de anime tem correlação com as maiores notas, mas mesmo assim vale a pena verificar quais são os gêneros de animes mais feitos e veremos isso graficamente.

```python
dez_generos = pd.DataFrame(dados.genre.value_counts().head(10))
dez_generos
```

```python
plt.figure(figsize=(15,5))
plt.bar(dez_generos.index, dez_generos.genre)
plt.title('10 gêneros mais produzidos')
plt.xticks(rotation=90)
plt.show()
```

### Até aqui vimos que o formato TV e Ova são os formatos com mais animes produzidos, mas o formato de Filmes "Movie", consegue superar OVA em nota altas o que deveria ser o que as produtoras buscam.

### Será que os formatos mais produzidos são os tem a melhor média de nota?

```python
dados.type.value_counts()
```

```python
notas_tipos = dados.groupby('type').agg({'rating':'mean'}).rating.sort_values(ascending=False).reset_index()
notas_tipos
```

```python
plt.figure(figsize=(15,5))
plt.bar(notas_tipos.type, notas_tipos.rating)
plt.title('Média de nota por tipo de anime')
plt.xticks(rotation=90)
plt.show()
```

### O formato para TV possui sim a melhor média porém tem pouco diferença do formato especial e OVA

### Média de notas por quantidade de episódios 

```python
nota_ep = dados.groupby('episodes').agg({'rating':'mean'}).reset_index()
nota_ep
```

```python
plt.figure(figsize=(15,5))
plt.scatter(nota_ep.episodes, nota_ep.rating)
plt.title('Média de notas por quantidade de episódio')
plt.show()
```

### No que podemos observar acima, o número de episódios não parece influenciar nas notas dos animes, pois animes com poucos episódios ou muito episódios podem ter notas altas, ou baixas.
